```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import re
import time 
import pandas as pd
import random
```

```python
driver = webdriver.Chrome(executable_path="./Driver/chromedriver.exe")
```

```python
urls = [
    "https://www.redfin.com/city/18142/FL/Tampa",
    "https://www.redfin.com/city/11458/FL/Miami",
    "https://www.redfin.com/city/13655/FL/Orlando"
]

home_listings= []
for url in urls:
    driver.get(url)
    time.sleep(random.randint(2,8))
    links = driver.find_elements_by_xpath("//a[@href]")
    for link in links:
        listing = link.get_attribute("href")
        if re.sub('https://www.redfin.com/city/\d{5}/', '', url) in listing:
            if listing not in home_listings and listing.find("/home/") != -1: 
                home_listings.append(listing)
                print(listing)

                
#add link to fetch data
#driver.get("https://www.redfin.com/city/18142/FL/Tampa")
home_listings
```

```python
#purchase price of property
driver.get("https://www.redfin.com/FL/Tampa/6626-Citrus-Creek-Ln-33625/home/142947944")
time.sleep(2)

price = driver.find_element_by_xpath("//div[contains(@class, 'price-section')]/div/div/span[2]").text
monthly_payment = driver.find_element_by_xpath("//div[contains(@class, 'CalculatorSummary')]/div/p").text
bedrooms = driver.find_element_by_xpath("//div[contains(@class, 'beds-section')]/div").text
baths = driver.find_element_by_xpath("//div[contains(@class, 'baths-section')]/div").text
square_feet = driver.find_element_by_xpath("//div[contains(@class, 'sqft-section')]/span").text
price, monthly_payment, bedrooms, baths, square_feet
```

```python
prices = []
monthly_payments = []
bedrooms = []
baths = []
square_feet = []
street_address = []
city = []

for listing in home_listings:
    driver.get (listing)
    time.sleep(random.randint(2,8))
    price = driver.find_element_by_xpath("//div[contains(@class, 'price-section')]/div/div/span[2]").text
    monthly_payment = driver.find_element_by_xpath("//div[contains(@class, 'CalculatorSummary')]/div/p").text
    bedrooms = driver.find_element_by_xpath("//div[contains(@class, 'beds-section')]/div").text
    baths = driver.find_element_by_xpath("//div[contains(@class, 'baths-section')]/div").text
    square_feet = driver.find_element_by_xpath("//div[contains(@class, 'sqft-section')]/span").text
    street_address = driver.find_element_by_xpath("//div[contains(@class, 'street-address')]/span").text
    city = driver.find_element_by_xpath("//h1[contains(@class, 'homeAddress-variant')]/div[2]").text
    prices.append (price)
    monthly_payments.append (monthly_payment)

df = pd.DataFrame({
     "price": prices, 
     "monthy_payment": monthly_payments
     "bedrooms": bedrooms
     "baths": baths
     "square_feet": square_feet
     "street_address": street_address
     "city": city
})
df
```
